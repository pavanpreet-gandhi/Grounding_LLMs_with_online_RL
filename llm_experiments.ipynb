```python
import warnings
from pprint import pprint
from score_ollama_llm import score_ollama_llm
```

```python
warnings.filterwarnings("ignore")
```

```python
llms = ["smollm2:135m", "smollm2:360m", "smollm2:1.7b", "llama3.2:1b", "llama3.2:3b", "qwen:0.5b", "qwen:1.8b", "qwen:4b"]
rewards_all = {}
invalid_actions_all = {}
for llm in llms:
    print(f"Scoring {llm}")
    rewards, invalid_actions = score_ollama_llm(model_name=llm, num_episodes=20, max_invalid_per_step=5, verbose=True)
    rewards_all[llm] = rewards
    invalid_actions_all[llm] = invalid_actions
```

```python
print("Rewards:")
pprint(rewards_all)
print()
print("Invalid actions:")
pprint(invalid_actions_all)
```

```python
import json

# Save rewards_all
with open('rewards_all.json', 'w') as f:
    json.dump(rewards_all, f, indent=4)

# Save invalid_actions_all
with open('invalid_actions_all.json', 'w') as f:
    json.dump(invalid_actions_all, f, indent=4)

print("Results saved to rewards_all.json and invalid_actions_all.json")
```

```python
rewards_all = json.load(open('rewards_all.json'))
invalid_actions_all = json.load(open('invalid_actions_all.json'))

import matplotlib.pyplot as plt

# boxplot for rewards with scatter plot overlay
fig, ax = plt.subplots()
ax.boxplot(rewards_all.values())
for i, (llm, rewards) in enumerate(rewards_all.items()):
    x = [i + 1] * len(rewards)
    ax.scatter(x, rewards, alpha=0.5)
ax.set_xticklabels(rewards_all.keys(), rotation=45)
ax.set_ylabel("Rewards")
plt.title("Rewards for different LLMs")
plt.show()

# print sum of invalid actions for each LLM
print("Sum of invalid actions for each LLM:")
for llm, invalid_actions in invalid_actions_all.items():
    print(f"{llm}: {sum(invalid_actions)}")
```
